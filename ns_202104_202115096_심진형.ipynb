```python
import chardet

# 'rb'는 Read Binary의 약자로, 해석하지 않은 날것의 데이터를 읽겠다는 뜻입니다.
with open('남산도서관 장서 대출목록 (2021년 04월).csv', 'rb') as f:
    raw_data = f.read(10000) # 파일의 앞부분 1만 바이트만 읽어서 검사
    result = chardet.detect(raw_data)

print(result)
```
